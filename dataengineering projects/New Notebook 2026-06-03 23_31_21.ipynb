Temperature and Severity Alerts

```python
from pyspark.sql.functions import when

alerts_df = clean_df.filter(
    (col("heart_rate") > 120) |
    (col("heart_rate") < 50) |
    (col("spo2") < 90) |
    (col("temperature") > 38.5)
).withColumn(
    "severity",
    when(col("spo2") < 85, "CRITICAL")
    .when(col("heart_rate") > 140, "HIGH")
    .otherwise("MEDIUM")
)
```
