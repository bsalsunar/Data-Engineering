```python
%sql
Create table if not exists emp(
  emp_id int,
  emp_name string,
  salary int,
  department string,
  hire_date date
);
```

```python
%sql
Insert overwrite emp values
(1,'ram',2000,'it','2023-01-10'),
(2,'sam',3000,'hr','2023-01-11'),
(3,'john',4000,'hr','2023-01-12'),
(4,'jane',5000,'it','2023-01-13'),
(5,'joe',6000,'hr','2023-01-14');
```

```python
%sql

select * from emp;
```

```python
%sql
select department,sum(salary) as totalsalary from emp
GROUP by department;
```

```python
%sql

select department,count(*) as emplyoenumbr from emp
group by department;
```

```python
%sql
select * from emp
where hire_date = (select max(hire_date) from emp);
```

```python
%sql
select *from(
  select *,
  row_number() over (partition by department order by salary desc) as rank
  from emp
)tem
where rank<=2


```
