```python
-- =========================================
-- SNOWFLAKE COMMON COMMANDS 
-- =========================================

-- =========================================
-- DATABASE COMMANDS
-- =========================================

CREATE DATABASE sales_db;

SHOW DATABASES;

USE DATABASE sales_db;

DROP DATABASE sales_db;


-- =========================================
-- SCHEMA COMMANDS
-- =========================================

CREATE SCHEMA finance_schema;

SHOW SCHEMAS;

USE SCHEMA finance_schema;

DROP SCHEMA finance_schema;


-- =========================================
-- TABLE COMMANDS
-- =========================================

CREATE TABLE employees (
    id INT,
    name STRING,
    salary NUMBER
);

SHOW TABLES;

DESC TABLE employees;

ALTER TABLE employees ADD COLUMN dept STRING;

ALTER TABLE employees RENAME TO emp;

DROP TABLE employees;


-- =========================================
-- DATA MANIPULATION COMMANDS (DML)
-- =========================================

INSERT INTO employees VALUES (1, 'John', 5000);

SELECT * FROM employees;

SELECT name, salary
FROM employees
WHERE salary > 4000;

UPDATE employees
SET salary = 6000
WHERE id = 1;

DELETE FROM employees
WHERE id = 1;

TRUNCATE TABLE employees;


-- =========================================
-- MERGE COMMAND
-- =========================================

MERGE INTO target t
USING source s
ON t.id = s.id
WHEN MATCHED THEN
    UPDATE SET t.name = s.name
WHEN NOT MATCHED THEN
    INSERT VALUES (s.id, s.name);


-- =========================================
-- WAREHOUSE COMMANDS
-- =========================================

CREATE WAREHOUSE compute_wh
WITH WAREHOUSE_SIZE = 'SMALL';

SHOW WAREHOUSES;

USE WAREHOUSE compute_wh;

ALTER WAREHOUSE compute_wh RESUME;

ALTER WAREHOUSE compute_wh SUSPEND;

DROP WAREHOUSE compute_wh;


-- =========================================
-- STAGE COMMANDS
-- =========================================

CREATE STAGE my_stage;

LIST @my_stage;

REMOVE @my_stage/file.csv;

DROP STAGE my_stage;


-- =========================================
-- FILE FORMAT COMMANDS
-- =========================================

CREATE FILE FORMAT csv_format
TYPE = CSV
SKIP_HEADER = 1;

SHOW FILE FORMATS;


-- =========================================
-- FILE LOADING COMMANDS
-- =========================================

PUT file://data.csv @my_stage;

COPY INTO employees
FROM @my_stage/data.csv
FILE_FORMAT = (TYPE = CSV);

COPY INTO employees
FROM @my_stage
FILE_FORMAT = csv_format;


-- =========================================
-- VIEW COMMANDS
-- =========================================

CREATE VIEW emp_view AS
SELECT * FROM employees;

SHOW VIEWS;

DROP VIEW emp_view;


-- =========================================
-- USER & ROLE COMMANDS
-- =========================================

CREATE USER john
PASSWORD = 'Password123';

CREATE ROLE analyst;

GRANT ROLE analyst TO USER john;

GRANT SELECT ON TABLE employees TO ROLE analyst;

REVOKE ROLE analyst FROM USER john;

DROP USER john;

DROP ROLE analyst;


-- =========================================
-- SNOWPIPE COMMANDS
-- =========================================

CREATE PIPE my_pipe
AS
COPY INTO employees
FROM @my_stage;

ALTER PIPE my_pipe REFRESH;

SHOW PIPES;

DROP PIPE my_pipe;


-- =========================================
-- STREAM COMMANDS
-- =========================================

CREATE STREAM emp_stream
ON TABLE employees;

SHOW STREAMS;

DROP STREAM emp_stream;


-- =========================================
-- TASK COMMANDS
-- =========================================

CREATE TASK my_task
WAREHOUSE = compute_wh
SCHEDULE = '1 MINUTE'
AS
INSERT INTO log_table VALUES (CURRENT_TIMESTAMP);

ALTER TASK my_task RESUME;

ALTER TASK my_task SUSPEND;

SHOW TASKS;

DROP TASK my_task;


-- =========================================
-- TIME TRAVEL COMMANDS
-- =========================================

SELECT * FROM employees
AT (TIMESTAMP => '2026-05-20 10:00:00');

UNDROP TABLE employees;


-- =========================================
-- CLONING COMMANDS
-- =========================================

CREATE TABLE emp_clone CLONE employees;

CREATE DATABASE sales_clone CLONE sales_db;


-- =========================================
-- TRANSACTION COMMANDS
-- =========================================

BEGIN;

COMMIT;

ROLLBACK;


-- =========================================
-- UTILITY COMMANDS
-- =========================================

SELECT CURRENT_USER();

SELECT CURRENT_ROLE();

SELECT CURRENT_DATABASE();

SELECT CURRENT_SCHEMA();

SELECT CURRENT_WAREHOUSE();

SELECT CURRENT_TIMESTAMP();

SHOW PARAMETERS;

SHOW GRANTS;

SHOW FUNCTIONS;

SHOW PROCEDURES;


-- =========================================
-- QUERY HISTORY
-- =========================================

SELECT *
FROM TABLE(INFORMATION_SCHEMA.QUERY_HISTORY());


-- =========================================
-- STORAGE INTEGRATION
-- =========================================

CREATE STORAGE INTEGRATION s3_int
TYPE = EXTERNAL_STAGE
STORAGE_PROVIDER = S3
ENABLED = TRUE;


-- =========================================
-- EXTERNAL STAGE
-- =========================================

CREATE STAGE ext_stage
URL='s3://mybucket/data/'
STORAGE_INTEGRATION = s3_int;


-- =========================================
-- EXTERNAL TABLE
-- =========================================

CREATE EXTERNAL TABLE ext_table (
    id INT,
    name STRING
)
LOCATION=@ext_stage
FILE_FORMAT=(TYPE=PARQUET);


-- =========================================
-- COMMON SHOW COMMANDS
-- =========================================

SHOW DATABASES;

SHOW SCHEMAS;

SHOW TABLES;

SHOW VIEWS;

SHOW STAGES;

SHOW USERS;

SHOW ROLES;

SHOW WAREHOUSES;

SHOW PIPES;

SHOW STREAMS;

SHOW TASKS;
```
