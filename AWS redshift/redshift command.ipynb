```python
%sql


CREATE TABLE orders (
    order_id INT,
    customer_id INT,
    order_date DATE
)
DISTKEY(customer_id)
SORTKEY(order_date);


COPY customers
FROM 's3://mybucket/customers.csv'
IAM_ROLE 'arn:aws:iam::123456789012:role/MyRedshiftRole'
CSV
IGNOREHEADER 1;

UNLOAD ('SELECT * FROM customers')
TO 's3://mybucket/output/customers_'
IAM_ROLE 'arn:aws:iam::123456789012:role/MyRedshiftRole'
CSV;
```
