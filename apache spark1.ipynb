```python
%python
#creating raw data
data= [
    (1,'phone',2000),
    (2,'laptop',4000),
    (3,'charger',500),
    (4,'mouse',100),
    (5,'keyboard',200)
    
]
#defining data columns
columns=["order_id","Products","Price"]

df=spark.createDataFrame(data,columns)

#apply filter and select
result=df.filter(df.Price>2000).select("Products","Price")

display(result)
#show table formats in databricks

```

```python
%python
#grouped by aggregration

#find total sales per product

from pyspark.sql.functions import sum
data1 = [
    (1,'charger',2000),
    (2,'mouse',4000),
    (3,'charger',500),
    (4,'mouse',100),
    (5,'keyboard',200)
    
]
#defining data columns
columns1=["order_id","Products","Price1"]

df=spark.createDataFrame(data1,columns1)

result=df.groupby("Products").agg(sum("price1").alias("tatal_sales"))
display(result)
```

```python
%python

#sorting data- show all orders sorted by price 

result=df.orderBy(df.Price1.desc)
display(result)
```

```python
%python
from pyspark.sql.functions import col
#create any new column - add a new column discount_price = price-100

df=spark.createDataFrame(data1,columns1)
new_result = df.withColumn("discounted_price", col("Price1")-100)
display(new_result)
```

```python
%python

#to remove duplucates
df=spark.createDataFrame(data1,columns1)
drop_df=df.dropDuplicates(["Products"])
display(drop_df)
```
