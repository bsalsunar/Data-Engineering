```python
%sql
-- having aftre grouping
--where before grouping
-- find departments with average salary greater than 55000

select department ,avg(salary) as avg_salary
from emp1
group by department
having avg(salary) >55000;
    
```

```python
%sql
-- we have to find department with more than 2 employees

SELECT department, COUNT(*) AS num_employees
FROM emp1
GROUP BY department
HAVING COUNT(*) > 2;
    
```

```python
%sql
-- Find Departments where
--average salary > 55000
--only include IT and Finance

select department,avg(salary) as avg_salary
from emp1
where department in ('IT','Finance')
group by department
having avg_salary >55000;
```

```python
%sql
CREATE TABLE IF NOT EXISTS employee (

  emp_id INT,

  emp_name STRING,

  department_id INT
);

CREATE TABLE IF NOT EXISTS departments (

  department_id INT,

  department_name STRING

);
```

```python
%sql 
select employee ,departments
```

```python
%sql
select employee.emp_name, departments.department_name 
from employee 
inner join departments 
on employee.department_id = departments.department_id;
```
