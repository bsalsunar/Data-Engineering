AThena validation for covid analytics 

athena silver stage validation 

```python
%sql
CREATE EXTERNAL TABLE IF NOT EXISTS covid_silver_db.cases_standardized (
  full_date date,
  state_code string,
  state_name string,
  cases_cum bigint,
  deaths_cum bigint,
  year int,
  month int,
  day int
)
PARTITIONED BY (state_code string, year int, month int, day int)
STORED AS PARQUET
LOCATION 's3://YOUR_BUCKET/covid/silver/cases_standardized/';
MSCK REPAIR TABLE covid_silver_db.cases_standardized;

-- Testing
CREATE EXTERNAL TABLE IF NOT EXISTS covid_silver_db.testing_standardized (
  full_date date,
  state_code string,
  state_name string,
  tests_total_cum bigint,
  tests_pos_cum bigint,
  tests_neg_cum bigint,
  year int, month int, day int
)
PARTITIONED BY (state_code string, year int, month int, day int)
STORED AS PARQUET
LOCATION 's3://YOUR_BUCKET/covid/silver/testing_standardized/';
MSCK REPAIR TABLE covid_silver_db.testing_standardized;

```

Athena for GOld stage validation** 

```python
%sql
CREATE DATABASE IF NOT EXISTS covid_gold_db;

CREATE EXTERNAL TABLE IF NOT EXISTS covid_gold_db.dim_date (
  date_id int,
  full_date date,
  year int,
  month int,
  day int,
  dow int,
  is_weekend boolean
)
STORED AS PARQUET
LOCATION 's3://covid-19-analytics-project2/covid/gold/dim_date/';

CREATE EXTERNAL TABLE IF NOT EXISTS covid_gold_db.dim_state (
  state_code string,
  state_name string
)
STORED AS PARQUET
LOCATION 's3://covid-19-analytics-project2/covid/gold/dim_state/';

CREATE EXTERNAL TABLE IF NOT EXISTS covid_gold_db.fact_cases_state_daily (
  date_id int,
  state_code string,
  cases_cum bigint,
  deaths_cum bigint,
  new_cases int,
  new_deaths int
)
PARTITIONED BY (state_code string)
STORED AS PARQUET
LOCATION 's3://covid-19-analytics-project2/covid/gold/fact_cases_state_daily/';
MSCK REPAIR TABLE covid_gold_db.fact_cases_state_daily;

CREATE EXTERNAL TABLE IF NOT EXISTS covid_gold_db.fact_testing_state_daily (
  date_id int,
  state_code string,
  tests_total_cum bigint,
  tests_pos_cum bigint,
  tests_neg_cum bigint,
  new_tests int,
  positivity_rate double
)
PARTITIONED BY (state_code string)
STORED AS PARQUET
LOCATION 's3://covid-19-analytics-project2/covid/gold/fact_testing_state_daily/';
MSCK REPAIR TABLE covid_gold_db.fact_testing_state_daily;
```
